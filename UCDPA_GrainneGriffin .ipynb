# EXPLORATORY DATA ANALYSIS

# Analysis of Failed Vs Successful Rocket Missions

```python
# Number of Successful vs Fauiled Rocket Missions since 1957

# Graph 1
sns.set(style="darkgrid")
plt.title("Success v Failure Rocket Missions")
ax = sns.countplot(x="Status Mission", data=space_df, palette=('Red','Blue'))

# can see successful missions outcome far greater than failures and partial failures
# their have been no prelaunch failures recorded
```

```python
# Number of Success vs Failure for SpaceX Only Missions

# Graph 2
sns.set(style="darkgrid")
plt.title("Number of Success vs Failure for SpaceX Only Missions")
ax = sns.countplot(x="Mission Outcome", data=spacex_df, palette=('Red','Blue'))

# can see successful missions outcome far greater than failures
```

```python
# Percentage of Rockets currently either Retired or Active?

# Graph 3
plt.figure(figsize=(12,6))
plt.title("Percentage of Rockets currently Retired or Active")
g = plt.pie(space_df["Status Rocket"].value_counts(), explode=(0.025,0.025), labels=space_df["Status Rocket"].value_counts().index, colors=['red','blue'],autopct='%1.1f%%', startangle=180);
plt.legend()
plt.show()

# illustrating the percentage difference of currently active missions compared to that which are retired
# Retired missions by far over weigh the actives with a value of 81.7% 
```

```python
# Count of Rocket Missions across Location

# Graph no. 4
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="Location", data=space_df, palette=("Set3"), order=space_df['Location'].value_counts().index[0:20])
plt.title('Count of Rocket Missions across Location')
plt.show()

# The largest count of Rocket missiles launched are within Kazakhstan
# After this the second, third and fourth largest is within Russia 
# the USA comes in first at 9th largest
```
